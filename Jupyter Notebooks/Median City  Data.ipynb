```python
median_US_households = ["Married Households", "Male Unmarried Household", "Female Unmarried Household", "Nonfamily Household"]
number_of_households = [median_married_US, median_male_US, median_Female_US, median_single_US]
x_axis = np.arange(len(number_of_households))
```

```python
plt.bar(x_axis, number_of_households, color="b", align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, median_US_households)
plt.xticks(x_axis, rotation=90)
plt.title("Median US households")
plt.ylabel("Population of households")
plt.show()
```

```python
state_medain_households = ["Married Households", "Male Unmarried Household", "Female Unmarried Household", "Nonfamily Household"]
number_of_state_households = [state_median, state_median_male, state_median_female, state_median_single]
x_axis = np.arange(len(number_of_state_households))
```

```python
plt.bar(x_axis, number_of_state_households, color="r", align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, state_medain_households)
plt.xticks(x_axis, rotation=90)
plt.title(f"Median {state} households")
plt.ylabel(f"Population of {state} households")
plt.show()
```

```python
median_US_households = ["Married Households", "Male Unmarried Household", "Female Unmarried Household", "Nonfamily Household"]
number_of_households = [median_married_US, median_male_US, median_Female_US, median_single_US]
state_medain_households = ["Married Households", "Male Unmarried Household", "Female Unmarried Household", "Nonfamily Household"]
number_of_state_households = [state_median, state_median_male, state_median_female, state_median_single]
city_medain_households = ["Married Households", "Male Unmarried Household", "Female Unmarried Household", "Nonfamily Household"]
number_of_city_households = [city_df['Married Households'].item(), city_df['Male Unmarried Household'].item(),
                                 city_df['Female Unmarried Household'].item(), city_df['Nonfamily Household'].item()]
x_axis = np.arange(len(number_of_households))

#Create the first bar chart
plt.bar(x_axis+0.0, number_of_households, width=0.2)

#Create the second bar chart
plt.bar(x_axis+0.6, number_of_state_households, width=0.2)

#Create the third bar chart
plt.bar(x_axis+0.3, number_of_city_households, width=0.2)

#Label the x-axis
plt.xlabel('Household Type')

plt.xticks(tick_locations, state_medain_households)
plt.xticks(x_axis, rotation=45)

plt.legend(['number_of_US_households', f'number_of_{state}_households', f'number_of_{city}_households'])

#Show the plot
plt.show()
```
