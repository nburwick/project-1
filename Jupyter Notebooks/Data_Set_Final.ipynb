```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_path = "../Outputs/Main_Data_Set_Cleaned.csv"
csv2_path = "../Outputs/Annual Census by City and Age.csv"
```

```python
# Import the CSV into a pandas DataFrame
MainDF = pd.read_csv(csv_path)
CensusAgeDF = pd.read_csv(csv2_path)
```

```python
MainDF.head(3)
```

```python
CensusAgeDF.head()
```

```python
FinalDF = pd.merge(MainDF, CensusAgeDF, on=["City", "State"])
```

```python
FinalDF
```

```python
FinalDF = FinalDF[['City', 'State','% Age Over 19','# Breweries in City','2021 Pop. Estimate']]
```

```python
FinalDF
```

```python
# Reorder Columns 
FinalDF = FinalDF[['City','State','2021 Pop. Estimate', '% Age Over 19','# Breweries in City']]
```

```python
FinalDF.drop_duplicates(inplace=True)
```

```python
FinalDF['People Per Brewery'] = (FinalDF['2021 Pop. Estimate'] * FinalDF['% Age Over 19']) / FinalDF['# Breweries in City']
FinalDF
```

```python
FinalDF = FinalDF.sort_values(['People Per Brewery'],ascending=False)[:10]
FinalDF
```

```python
# Set x axis and tick locations
x_axis = FinalDF['City']
```

```python
ticks = [ x for x in x_axis]
```

```python
plt.bar(x_axis , FinalDF["People Per Brewery"], color = 'red', alpha=.75, align="center")
plt.xticks(np.arange(len(FinalDF)), rotation=45)
plt.xlabel("Cities in US")
plt.ylabel("Number of People over Age 19 Per Brewery")
plt.title("People over Age 19 Per Brewery in Cities (US)")
plt.show()
plt.tight_layout()
```
