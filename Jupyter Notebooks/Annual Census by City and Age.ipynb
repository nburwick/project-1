```python
# Import Dependencies
import pandas as pd
import os
```

```python
csv_path = "../Resources/CensusAgeSex2021.csv"
csv2_path = "../Resources/StateAbb.csv"
csv3_path = "../Resources/StateAbb.csv"
```

```python
censusAge = pd.read_csv(csv_path)
stateAbbr = pd.read_csv(csv2_path)
stateAbbrDF = pd.read_csv(csv3_path)
```

```python
# Necessary Columns
censusAge_DF = censusAge[['NAME', 'AGE04_TOTAL', 'AGE59_TOTAL','AGE1014_TOTAL','AGE1519_TOTAL','POPESTIMATE', 'DATE']]
```

```python
# Reorder Columns 
#censusAge_DF = censusAge_DF[['NAME','DATE','POPESTIMATE','AGE18PLUS_TOTAL', 'AGE18PLUS_MALE']]
```

```python
# The key for DATE is as follows:
# 1 = 4/1/2020 population estimates base
# 2 = 7/1/2020 population estimate
# 3 = 7/1/2021 population estimate
# Only show population estimate collected on 7/1/2021 which is 3
censusAge_DF = censusAge_DF.drop(censusAge_DF[censusAge_DF['DATE'] < 3].index)
```

```python
# Get the number/percentage of population that are older than 20
over19 = censusAge_DF['POPESTIMATE'] - censusAge_DF['AGE04_TOTAL'] - censusAge_DF['AGE59_TOTAL'] - censusAge_DF['AGE1014_TOTAL']  - censusAge_DF['AGE1519_TOTAL']  
over19percentage = over19 / censusAge_DF['POPESTIMATE']
```

```python
# Add the columns
censusAge_DF['Age Over 19'] = over19
censusAge_DF['% Age Over 19'] = over19percentage
```

```python
censusAge_DF
```

```python
# Extract City
censusAge_DF['City'] = censusAge_DF['NAME'].str[:-4]
# Extract State
censusAge_DF['State'] = censusAge_DF['NAME'].str[-2:]
```

```python
# Rename
stateAbbrDF = stateAbbrDF.rename(columns={"USPS Abbreviation":"State",})
stateAbbrDF.head()
```

```python
# Show only necessary columns
censusAge_DF = censusAge_DF[['City', 'State', '% Age Over 19']]
censusAge_DF.head(2)
```

```python
censusAge_DF = pd.merge(censusAge_DF, stateAbbrDF, on="State", how="left")
```

```python
censusAge_DF
```

```python
censusAge_DF = censusAge_DF[['City', 'State Name', '% Age Over 19']]
censusAge_DF = censusAge_DF.rename(columns={'State Name':'State',})
censusAge_DF
```

```python
# Push DataFrame to csv
censusAge_DF.to_csv(os.path.join('..', 'Outputs', 'Annual Census by City and Age.csv'), index=False)
```
