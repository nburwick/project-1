```python
# import dependancies
import pandas as pd
import numpy as np
import os
```

```python
# Read Resource csv files into script
df_1 = pd.read_csv(os.path.join('..','Resources', 'cost-of-living.csv'), delimiter=',', index_col=0, )
df_2 = pd.read_csv(os.path.join('..','Resources', 'cost-of-living_v2.csv'), delimiter=',' )
# Display Data Frames
print('Cost Of Living (V1):')
display(df_1)
print('Cost Of Living (V2):')
display(df_2)
```

```python
# Merge Data Frames
df_merged = pd.merge(df_1,df_2 , how='outer', indicator=True)
```

```python
# Get Current Column Names
cols = df_merged.columns.tolist()
```

```python
# Create List of Column Names based off of Documentation - https://www.kaggle.com/datasets/mvieira101/global-cost-of-living?select=cost-of-living.csv
col_names = ['City', 'Country', 'Avg Meal Inexp. Restaurant', 'Avg Meal (For 2) Mid-Range Restaurant ',
             'McMeal at McDonalds', 'Domestic Beer (Restaurant)', 'Imported Beer (Restaurant)', 'Cappuccino',
             'Coke/Pepsi (Restaurant)', 'Water (Restaurant)','Milk', 'Bread (Loaf)', 'Rice', 'Dozen Eggs','Cheese (Local)',
             'Chicken Fillets', 'Beef Round', 'Apples', 'Banana',
             'Oranges', 'Tomato', 'Potato', 'Onion', 'Lettuce', 'Water Bottle', 'Wine Bottle', 'Domestic Beer (Bottle)',
             'Imported Beer (Bottle)', 'Cigarettes (Marlboro - 20 cnt)', 'One-Way Ticket (Local)', 'Monthly Pass (Local)',
             'Taxi (Start)', 'Taxi 1km', 'Taxi 1-hr Wait', 'Gasoline (per Liter)', 'Volkswagen Golf', 'Toyota Corolla', 
             'Basic Utilities', '1 min PrePaid Phone', 'Internet', 'Fitness Club Membership', 'Tennis Court Rental', 'Cinema',
             'Preschool', 'Primary School', 'Pair of Jeans', 'Summer Dress', 'Nike Shoes', 'Leather Business Shoes',
             '1bd Apt. (City Center)', '1bd Apt. (Outside Center)', '3bd Apt. (City Center)', '3bd Apt. (Outside Center)',
             'Price/sqm Buy Apt. (City Center)', 'Price/sqm Buy Apt. (Outside Center)', 'Avg. Monthly Income (After Tax)', 'Avg. Mortgage Interest (%)',
             'Data Quality', 'Merge']
```

```python
# Zip Current and Future Column names into dictionary
renamed = dict(zip(cols, col_names))
```

```python
# Rename Columns
df_merged.rename(columns=renamed, inplace=True)
# Display Merged Data
display(df_merged)
```

```python
# Filter to United States
US_df = df_merged[df_merged['Country'] == 'United States']
```

```python
# Identify Any Duplicate Cities
dup_cities = US_df[US_df['City'].duplicated()]['City']
print(f"There were {len(dup_cities)} duplicates")
```

```python
# Explore Duplicate Data
dup_df = US_df[US_df['City'].isin(dup_cities)]
#Display Duplicate Data
display(dup_df.sort_values('City'))
```

```python
# Because the data seems to be a difference only between the original data sets
# we will take an average of the two duplicates
US_df_grouped = US_df.groupby('City').mean(numeric_only=True)

#Display Cleaned Data
display(US_df_grouped)
```

```python
# Push to csv for next portion of project
US_df_grouped.to_csv(os.path.join('..','Outputs', 'United_States_City_Data.csv'))
```
