```python
# Import Dependencies
import pandas as pd
import numpy as np
import requests
import os
import json
from pprint import pprint
```

```python
# Get Cities/Breweries from Output Files
cl_df = pd.read_csv(os.path.join('..', 'Outputs', 'United_States_City_Data.csv'), delimiter=',')
beer_df = pd.read_csv(os.path.join('..', 'Outputs', 'Popular Beers.csv'), delimiter=',')
```

```python
# Display Previews of dfs
print("Cost Of Living Data:")
display(cl_df.head())
print("Popular Beers:")
display(beer_df.head())
```

```python
# Merge Current Data into single DataFrame
found_cities = pd.merge(beer_df, cl_df, how='left', on='City', indicator=True)
display(found_cities)
```

```python
# Drop rows that were not found in both and keep only found rows
found_cities = found_cities[found_cities['_merge'] == 'both']

# Drop Unwanted Columns
found_cities.drop(labels=['Cinema', 'Leather Business Shoes','Cigarettes (Marlboro - 20 cnt)','Preschool',
                          'Primary School', 'Tennis Court Rental', 'Nike Shoes', 'Data Quality', 
                          '1 min PrePaid Phone', 'Cappuccino', 'McMeal at McDonalds', 'Volkswagen Golf', 
                          'Toyota Corolla', 'Avg. Mortgage Interest (%)' , 'Pair of Jeans', 'Summer Dress', 
                          'Fitness Club Membership', '_merge'], axis=1, inplace=True)
# Display New Data Preview
display(found_cities.head())

# Explore Data Further
print(found_cities.info())
```

```python
# Create List of Unique Cities After Merge and Cleaning
cities_ls = found_cities['City'].unique().tolist()
print(f"Found {len(cities_ls)} Unique Cities in Data.")
```

```python
# Create API Call
# Create Empty df_list to hold DataFrames from API Call
df_list = []

# Loop through cities and call API and append to df_list if there is data
for city in cities_ls:
    try:
        OBD_response = requests.get("https://api.openbrewerydb.org/breweries",params={'by_city': city}).json()
        OBD_df = pd.json_normalize(OBD_response)
        if not OBD_df.empty:
            OBD_df['# Breweries in City'] = int(OBD_df.shape[0])
            df_list.append(OBD_df)
    except:
        pass
```

```python
# Concat Data into single DataFrame
api_df = pd.concat(df_list, ignore_index=True)

# Display Preview of Data
display(api_df.head())
```

```python
# Explore API Data
print(api_df.info())
```

```python
# Create List of Brewery Names to find matches from API Call
breweries = found_cities['Brewery Name'].unique().tolist()
print(f"Found {len(breweries)} breweries in Data")
```

```python
# Loop through brewery names to correct data to match original dataset names
for brew in breweries:
    for index, row in api_df.iterrows():
        string_api = row['name'].strip()
        if string_api == brew[:len(string_api)]:
            api_df.loc[index,'name'] = brew
```

```python
# Merge API Data to Current Data Set
merged_df = pd.merge(found_cities, api_df, how='left', left_on=['Brewery Name'], right_on=['name'], indicator=True)

# Clean Data - Keep Found Rows and Drop Duplicates
merged_df = merged_df[merged_df['_merge'] == 'both'].drop(labels=['_merge'],axis=1).drop_duplicates(subset=['Brewery ID', 'City','State'])
merged_df['State'] = merged_df['state']


# Drop Unwanted Columns
merged_df.drop(labels=['id', 'name', 'street', 'address_2','address_3','city','state','county_province',
                       'postal_code', 'phone', 'updated_at', 'created_at'], axis=1, inplace=True)

# Rename Columns
merged_df.rename(columns={'longitude':'Lng', 'latitude': 'Lat',
                          'country': 'Country', 'brewery_type': 'Brewery Type',
                          'website_url': 'website'}, inplace=True)
# Preview Data
display(merged_df.head())

# Explore Data
print(merged_df.info())
```

```python
# Bring in US Census Data Population Estimates by City - 
pop_df = pd.read_excel(os.path.join('..', 'Resources','Annual Census by City.xlsx'), index_col=0, header=None, names=['Geographic Area','Base Pop. Estimate', '2020 Pop. Estimate', '2021 Pop. Estimate'], skiprows=4)

# Clean Data of Notes from bottom of file
pop_df.dropna(subset='Geographic Area', inplace=True)

# Preview Data
display(pop_df.head())
```

```python
# Split Geographic Area to City and State Columns
pop_df[['City','State']] = pop_df['Geographic Area'].str.split(' city, ', regex=True, expand=True)
pop_df.drop(labels='Geographic Area', axis=1, inplace=True)

# Reorder Columns
pop_df = pop_df[['City', 'State', 'Base Pop. Estimate', '2020 Pop. Estimate', '2021 Pop. Estimate']]

# Calculate Year Over Year Growth Rate (%)
pop_df['Pop. Growth (%)'] = round((pop_df['2021 Pop. Estimate'] / pop_df['2020 Pop. Estimate']) - 1, 2)* 100

# Convert dtypes for Pop. Estimate columns
pop_df[['Base Pop. Estimate', '2020 Pop. Estimate', '2021 Pop. Estimate']] = pop_df[['Base Pop. Estimate', '2020 Pop. Estimate', '2021 Pop. Estimate']].astype(int)

# Display Preview of Cleaned Census Data
display(pop_df)
```

```python
# Merge Census Data to Merged Data
main_df = pd.merge(merged_df, pop_df, how='inner', on=['City', 'State'])

# Sort df by Growth Rate (%)
main_df.sort_values('Pop. Growth (%)', ascending=False, inplace=True)

# Preview DataFrame
display(main_df.head())
```
